## 1. Voorbereiding en Opstelling

### 1.1. Bepaal de watermassa ($m_{\text{water}}$)

1.  Weeg het lege bekerglas met roerder nauwkeurig met de weegschaal. Noteer deze massa ($m_{\text{beker}}$).
2.  Vul het bekerglas met water (ongeveer **$200 \ \text{g}$**).
3.  Weeg het bekerglas met water. Noteer de totale massa ($m_{\text{totaal}}$).
4.  Bereken de massa van het water: $m_{\text{water}} = m_{\text{totaal}} - m_{\text{beker}}$.

### 1.2. Zet de opstelling klaar

* Plaats het bekerglas op de magnetische roerder.
* Plaats de thermometer en de $10 \ \Omega$ weerstand ondergedompeld in het water. Zorg dat ze elkaar niet raken.
* Sluit de weerstand aan op de voedingsbron (DC). De voedingsbron moet UIT staan. Zet de roerder aan op hoogste stand zodat de warmte verdeeld wordt.
* Sluit een multimeter (ingesteld als voltmeter) parallel aan op de weerstand om de spanning ($U$) te meten.

### 1.3. Elektrische instelling

* Stel de voedingsbron in op **$\mathbf{20 \ \text{V}}$** ($U = \sqrt{40 \ \text{W} \cdot 10 \ \Omega}$). Dit levert het maximum veilige vermogen van $40 \ \text{W}$.

---

## 2. Dataverzameling

### 2.1. Initialisatie

1.  Start de roerder op een matige snelheid.
2.  Wacht $\mathbf{5 \ \text{minuten}}$ totdat de temperatuur stabiel is.
3.  Noteer deze als de starttemperatuur ($T_{\text{start}}$).

### 2.2. Start meting

1.  Zet de voedingsbron AAN en start de stopwatch exact tegelijkertijd.
2.  Noteer de werkelijke, gemeten spanning ($U$) op de multimeter.

### 2.3. Temperatuur meten

1.  Meet en noteer de temperatuur ($T$) elke $\mathbf{30 \ \text{seconden}}$ gedurende de verwarming.

### 2.4. Einde experiment

1.  Stop de meting na een totale tijd van $\mathbf{10 \ \text{minuten}}$ ($\mathbf{600 \ \text{s}}$).
2.  Zet onmiddellijk de voedingsbron UIT en stop de stopwatch.
3.  Noteer de finale temperatuur ($T_{\text{eind}}$) en de totale tijd ($t_{\text{eind}}$).

---

##  Veiligheidsinstructies

* De voedingsbron mag alleen AAN staan wanneer de weerstand volledig ondergedompeld is in het water.
* De roerder moet AAN staan om oververhitting van de weerstand te voorkomen en de warmte goed te verdelen.
* Raak de weerstand niet aan tijdens of direct na het experiment (i.v.m. heet oppervlak).
