# Introductie

Water heeft een enorme buffercapaciteit: je moet veel energie toevoegen om de temperatuur van water een graad te verwarmen. In dit practicum gaan we de soortelijke warmte van water bepalen door een bekende hoeveelheid water te verwarmen met een bekende hoeveelheid energie, en de temperatuurstijging te meten.

# Theorie

Warmte-energie $(Q)$ is de energie die wordt overgedragen als gevolg van een temperatuurverschil. Wanneer er water wordt verwarmd, wordt er warmte-energie toegevoegd, wat resulteert in een temperatuurstijging $(\Delta T)$. De relatie tussen deze grootheden wordt beschreven door de formule:$$Q = m \cdot c \cdot \Delta T$$waarin: 
- $Q$ de toegevoegde warmte-energie is ($J$).
- $m$ de massa van de stof is ($kg$).
- $\Delta T$ de temperatuurverandering is ($K$ of $^\circ C$).
- $c$ de soortelijke warmtecapaciteit van de stof is ($\text{J/(kg} \cdot K)$).

De soortelijke warmte ($c$) is een stofeigenschap die aangeeft hoeveel energie nodig is om de temperatuur van één kilogram van die stof met één graad Celsius ($1^\circ C$) of Kelvin ($1 K$) te verhogen. De referentiewaarde voor de soortelijke warmtecapaciteit ($c$) voor vloeibaar water bij standaardomstandigheden ($15^\circ C$) is:$$c_{water, theoretisch} \approx 4186 \text{ J/(kg} \cdot K)$$. In dit practicum leveren we een bekende hoeveelheid elektrische energie $(Q_{elektrisch} = P \cdot \Delta t)$ aan het water, meten we de massa $(m)$ en de temperatuurstijging $(\Delta T)$ over de tijd $(\Delta t)$, en lossen we vervolgens $c$ op.$$c = \frac{Q}{m \cdot \Delta T}$$

# Resultaten


````python
# Hier de data en de analyse
# watermassa berekening
m_bekerglas_met_roerstaafje_start = 0.1520 # kilogram
u_weegschaal = 0.0001 # kgram
m_bekerglas_met_roerstaafje_en_water_start = 0.3576 # kg
m_water_start = m_bekerglas_met_roerstaafje_en_water_start - m_bekerglas_met_roerstaafje_start
print("massa van water start:", m_water_start, "+/- 0.0001 kg")

# onzekerheden en initiele waarden
u_multimeter_voltage = 0.1 # V
u_temperatuur = 0.05 #K geschat
c_referentie = 4186  # J/(kg·K)
U = 19.90 # V
R = 10.0 # Ohm

# uiteindelijk stabiel gehouden op 5.06 V en toen verwarmd naar 19.90 V voor 10 min elke 30 sec gemeten.
# Plotten
file_path = r'C:\Users\Eline\Dropbox\Eline\Universiteiten\TUDelft\Applied Physics\BSc 2025-2026 jaar 1\Q2\IP2\practica.portfolio\c_water\rawdata.soortelijke.warmte.Eline.csv'

data = np.loadtxt(file_path, delimiter=',', skiprows=1) 
t = data[:, 0]
T = data[:,1]

# berekening vermogen P
P = (U**2) / R
print(f"Elektrisch vermogen (P):")
print(f"   {U**2} / {R} = {P:.2f} Watt (W)")

# Berekening helling
(helling, intercept) = np.polyfit(t, T, 1)
T_fit = helling * t + intercept

print(f"Bepaling van de helling (ΔT/Δt) in het lineaire deel:")
print(f"   De helling (ΔT/Δt) is: {helling:.4f} °C/s")

# Theoretische trendlijn
T_0 = 25.1        # gemeten initiële temperatuur (°C)
P = (U**2) / R
# Theoretische Helling = P / (m * c)
helling_theoretisch = P / (m_water_start * c_referentie)
t_test = np.linspace(1, 600, 600) # Tijd van 0 tot 600 seconden, met stappen van 1 seconde
T_theoretisch = T_0 + helling_theoretisch * t_test

#plot
plt.figure(dpi=400)
plt.title('Vergelijking Experimentele en Theoretische Temperatuurstijging', fontsize=16)
plt.plot(t,T, '+', label='Gemeten Data ($T$ vs $t$)')
plt.plot(t, T_fit, '--', label=f'Beste Rechte Lijn: Helling = {helling:.4f} K/s', color='red')
plt.plot(t_test, T_theoretisch, '-', label=f'Theoretische Trendlijn', color='orange', linewidth=2)
plt.legend(loc='lower right', fontsize=12)
plt.xlabel("$t$ (s)")
plt.ylabel("$T$ $(^\circ C)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(bottom=0, top=1.2*max(T))
plt.tight_layout()
plt.show()

# Berekening soortelijke warmte water
c_experimenteel = (P / m_water_start) * (1 / helling)

# Berekening van de relatieve fout: (|c_exp - c_ref| / c_ref) * 100%
relatieve_fout_percentage = (abs(c_experimenteel - c_referentie) / c_referentie) * 100
print(f"Berekening van de soortelijke warmte van water (c):")
print(f"   c experimenteel: {c_experimenteel:.2f} J/(kg·K)")
print(f"   Referentiewaarde c: {c_referentie} J/(kg·K)")
print(f"   Relatieve Fout: |{c_experimenteel:.2f} - {c_referentie}| / {c_referentie} * 100%")
print(f"   Relatieve Foutpercentage: {relatieve_fout_percentage:.2f}%")
````

Omdat de formule $c = \frac{P}{m} \cdot \frac{1}{\text{helling}}$ een omgekeerd evenredig verband heeft met de helling, leidt een lagere helling tot een hogere $c_{experimenteel}$. De extra energie die verloren ging aan de omgeving is nu ten onrechte toegeschreven aan het water, wat resulteert in een overschatting van de soortelijke warmte. Er zal dus veel warmte van het verwarmingselement verloren zijn gegaan aan de lucht, de maatbeker en het roerstaafje.
